```python
#Task 1.1: Data Loading and Exploration
import numpy as np
import pandas as pd
import glob

from google.colab import files, drive
#uploaded = files.upload()
drive.mount('/content/drive')
dfPath=""
for file in glob.glob("/content/drive/MyDrive/**/StudentsPerformance.csv", recursive=True):
    dfPath=(file)

# Load the dataset
df=pd.read_csv(dfPath)
# print(df)

# Your tasks:
# 1. Display first 10 rows
# 2. Show dataset shape and info()
# 3. Display unique values in each categorical column
# 4. Show basic statistics for all score columns


# Your code here:
print("************** Display first 10 rows ************** ")
print(df[0:10])
print("************** Shapes are printed to know size of the dataset **************")
print(df.shape)
print("************** Info printed shows the summary of the data **************")
print(df.info())
print("************** Unique values in each categorical column **************")
categorical_cols=df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
  print(col,df[col].unique())
print("************** Basic statistics for all score columns **************")
df.describe(include='all')
```

```python
#Task 1.2: NumPy Array Operations
# Convert score columns to numpy arrays
math_scores = df['math score'].values
reading_scores = df['reading score'].values
writing_scores = df['writing score'].values

# Your tasks:
# 1. Calculate mean, median, std, min, max for each subject using NumPy
# 2. Find total students who scored above 80 in math
# 3. Find total students who scored below 50 in any subject
# 4. Calculate the overall average score across all three subjects

# Expected output example:
# Math - Mean: 66.1, Median: 66.0, Std: 15.2
# Students with math > 80: 132
# Students with any score < 50: 178

# Your code here:
print("************** mean, median, std, min, max for each subject **************")
print(f"Math Scores Mean: {np.mean(math_scores)}")
print(f"Math Scores Median: {np.median(math_scores)}")
print(f"Math Scores STD: {np.std(math_scores)}")
print(f"Math Scores Min: {np.min(math_scores)}")
print(f"Math Scores Max: {np.max(math_scores)}")

print(f"Reading Scores Mean: {np.mean(reading_scores)}")
print(f"Reading Scores Median: {np.median(reading_scores)}")
print(f"Reading Scores STD: {np.std(reading_scores)}")
print(f"Reading Scores Min: {np.min(reading_scores)}")
print(f"Reading Scores Max: {np.max(reading_scores)}")

print(f"Writing Scores Mean: {np.mean(writing_scores)}")
print(f"Writing Scores Median: {np.median(writing_scores)}")
print(f"Writing Scores STD: {np.std(writing_scores)}")
print(f"Writing Scores Min: {np.min(writing_scores)}")
print(f"Writing Scores Max: {np.max(writing_scores)}")

print("************** Total students who scored above 80 in math **************")
total_80=math_scores>80
# print(total_80)
print(f"Above 80 in math, there are: {np.count_nonzero(total_80)} persons")

print("************** Total students who scored below 50 in any subject **************")
total_bwlow_50_math=math_scores<50
total_bwlow_50_reading=reading_scores<50
total_bwlow_50_writing=writing_scores<50

print(f"Below 50 in any subject, there are: {np.count_nonzero(total_bwlow_50_math)+ np.count_nonzero(total_bwlow_50_reading)+ np.count_nonzero(total_bwlow_50_writing)} persons in total")

print("************** Overall average score across all three subjects **************")
average= (f"Overall average score across all three subjects is: {(np.sum(math_scores)+np.sum(reading_scores)+np.sum(writing_scores)) / (len(math_scores)+len(reading_scores)+len(writing_scores))}")
print(average)
```

```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
df_practice = df.copy()
print("Original missing values:", df_practice.isnull().sum())

# Step 2: Create missing data
np.random.seed(42)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50, replace=False)
missing_writing_idx = np.random.choice(df_practice.index, 30, replace=False)

# Continue the task...
# Your code here:

# print(missing_math_idx)
# print(missing_writing_idx)
# 2. Randomly set 50 math scores to np.nan
df_practice.loc[missing_math_idx, "math score"]=np.nan
# 3. Randomly set 30 writing scores to np.nan
df_practice.loc[missing_writing_idx, "writing score"]=np.nan

print("Original missing values added:", df_practice.isnull().sum())
# print(df_practice.isnull().sum())

# 4. Fill missing math scores with the MEDIAN math score
median_math_score=df_practice["math score"].median()
# print(median_math_score)
df_practice["math score"]=df_practice["math score"].fillna(median_math_score)
# print(df_practice.isnull().sum())

# 5. Fill missing writing scores with the MEAN writing score
mean_writing_score=df_practice["writing score"].mean()
# print(median_math_score)
df_practice["writing score"]=df_practice["writing score"].fillna(mean_writing_score)

# 6. Verify no missing values remain
print("No missing values :", df_practice.isnull().sum())
# print(df_practice.isnull().sum())

```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

# Your code here:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students
# numeric_cols = df_practice.select_dtypes(include=[np.number]).columns
numeric_cols=df_practice.loc[:, ["math score", "reading score", "writing score"]].columns
group_by_gender=df_practice.groupby("gender")[numeric_cols]
print(" Average scores in all subjects \n",(group_by_gender.mean().sum(axis=1))/3)
print(" Count of students \n",group_by_gender.size().sum())
# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

print(" Mean scores for each subject \n ", df_practice.groupby("test preparation course")[numeric_cols].mean())
subject_benefits=  df_practice.groupby("test preparation course")[numeric_cols].mean()

# subject_name=[for col in subject_benefits.columns]
# print([for col in ])
max_benifit=subject_benefits.max().max()
# cols=subject_benefits.columns[subject_benefits.max().max().index]
print("Most benefits of subject : ",subject_benefits.max().idxmax(), max_benifit)
# print()

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

print(" *****Group by parental level of education and calculate average math score for each edu level *****")
parentalLevelOfEducation=df_practice.groupby('parental level of education')
print(parentalLevelOfEducation[["math score"]].mean())
print(" *****Group by parental level of education and calculate Rank education levels by math performance *****")
print(parentalLevelOfEducation[["math score"]].mean().sort_values(by="math score",ascending=False))

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

print("****** Group by 'lunch' type ******")

group_by_lunch=df_practice.groupby("lunch")[numeric_cols]

print(group_by_lunch.mean())
print(group_by_lunch.size())

```
